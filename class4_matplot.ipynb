```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('garden.jpeg')
img2 = cv2.imread('pika.png')
```

```python
#img2 badi hai img se isiliye humne use crop kiya hai img ke according
img2 = img2[50:50+130,0:0+195,:]
```

```python
img.shape, img2.shape
```

```python
def showing(im):
    plt.imshow(im[:,:,::-1])
    plt.axis('off')
    plt.show()
```

```python
showing(img)
```

```python
showing(img2)
```

```python
def imshow(im):
    cv2.imshow('Preview',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
imshow(img)
imshow(img2)
```

```python
#IMAGE ARITHMETIC:-
```

```python
added = cv2.add(img,img2)
showing(added)
#hum kuch bhi value add karenge toh image bright hogi
```

```python
subtract = cv2.subtract(img,img2)
showing(subtract)
```

```python
#contrast badhane ke liye
multiply = cv2.multiply(img,1.1) #img ko 1.1 se multiply karenge
showing(multiply)
```

```python
img2_1 = cv2.multiply(img2,2*np.ones(img2.shape).astype('uint8'))
showing(img2_1)
```

```python
#brightness and contrast adjustment
# alpha for contrast
# beta for brightness
img2_2 = cv2.convertScaleAbs(img2,alpha=1.1,beta=30)
showing(img2_2)
```

```python
#image ko 20 pixels se contrast kam karna hai
img20_1 = cv2.subtract(img2,20*np.ones(img2.shape).astype('uint8'))
showing(img20_1)
```

```python
from skimage.transform import rotate
```

```python
img2_5 = rotate(img2,angle=45, resize=True)
showing(img2_5)
```

```python
img2_5 = rotate(img2,angle=-45, resize=True) #-45 for clockwise rotation
showing(img2_5)
```

```python
#CREATING AN ANIMATION
from time import sleep
repeat = True
while repeat:
    for i in range(360):
        i = rotate(img2,angle=i,resize = True)
        cv2.imshow("",i)
        key = cv2.waitKey(1)
        if key == ord('x'):
            repeat = False
            break
        sleep(0.1)
cv2.destroyAllWindows()        
```

```python
# RGB

# img_gray - Gray
# Red plane - img[:,:,0]
# green plane - img[:,:,1]
# blue plane - img[:,:,2]

# BGR

# img_gray - Gray
# Red plane - img[:,:,2]
# green plane - img[:,:,1]
# blue plane - img[:,:,0]
```

```python
img_gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
im_r = img2[:,:,2].copy()
im_g = img2[:,:,1].copy()
im_b = img2[:,:,0].copy()
```

```python
plt.subplot(1,5,1) #make a grid 1 row 5 col
plt.imshow(img2[:,:,::-1])
plt.axis('off')

plt.subplot(1,5,2) 
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1,5,3) 
plt.imshow(im_r, cmap='Reds')
plt.axis('off')

plt.subplot(1,5,4) 
plt.imshow(im_g, cmap='Blues')
plt.axis('off')

plt.subplot(1,5,5) 
plt.imshow(im_b, cmap='Greens')
plt.axis('off')
```
