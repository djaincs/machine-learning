```python
#jab bhi video capture karne me ya camera on karne me koi issue aa rhe ho toh poora code ek hi cell me le aana hai
import cv2
from time import sleep

vid = cv2.VideoCapture(0)

#to stop when we want just press x otherwise run infinitely
while(True):
    flag,img = vid.read()
    if flag:
        cv2.imshow('Preview',img)
        key = cv2.waitKey(1) #click image then wait for 1 sec then again click wait...
        if key == ord('x'):
            break
    else:
        break    
    sleep(0.1) #delay of 0.1 s    
vid.release()        #to close the camera after capturing
cv2.destroyAllWindows()
  
```

```python
#crop na karke ek particular area highlight karne ke liye ya uspar square banane ke liye
#to crop any image
import cv2
from time import sleep

vid = cv2.VideoCapture(0)

#to stop when we want just press x otherwise run infinitely
while(True):
    flag,img = vid.read()
    if flag:

        img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
        x1,y1,w,h = (200,400,200,200)

        cv2.rectangle(img, 
                      pt1 = (x1,y1), pt2 = (x1+w, y1+h),
                        color=(0,0,255), thickness=10)

        cv2.imshow('Preview',img)
        key = cv2.waitKey(1) #click image then wait for 1 sec then again click wait...
        if key == ord('x'):
            break
    else:
        break    
    sleep(0.1) #delay of 0.1 s    
vid.release()        #to close the camera after capturing
cv2.destroyAllWindows()
```

```python
#to detect face
import cv2
from time import sleep

fd = cv2.CascadeClassifier(
    cv2.data.haarcascades +
        'haarcascade_frontalface_default.xml'
)

vid = cv2.VideoCapture(0)

#to stop when we want just press x otherwise run infinitely
while(True):
    flag,img = vid.read()
    if flag:

        img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
        x1,y1,w,h = (200,400,200,200)

        faces = fd.detectMultiScale(img_gray, 1.1, 5)

        for x1,y1,w,h in faces:

            img_cropped = img[y1:y1+h, x1:x1+w, :]

            cv2.rectangle(img, 
                      pt1 = (x1,y1), pt2 = (x1+w, y1+h),
                        color=(0,255,0), thickness=5)

        cv2.imshow('Preview',img)
        key = cv2.waitKey(1) #click image then wait for 1 sec then again click wait...
        if key == ord('x'):
            break
    else:
        break    
    sleep(0.1) #delay of 0.1 s    
vid.release()        #to close the camera after capturing
cv2.destroyAllWindows()
```
