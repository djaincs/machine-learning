```python
cv2.COLOR_RGB2GRAY
```

```python
#TO CONVERT IMAGE INTO GRAY COLOR
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
ski.io.imshow(img_gray,)
```

```python
img_gray
```

```python
#TO CONVERT IMG INTO BINARY
a, img_binary = cv2.threshold(img_gray,50,255,cv2.THRESH_BINARY) #50,255 are threshold limits i.e. 50 se 255 ke beech me jitni
#bhi values aaegi use white kar dega or 50 se kam or 255 ke baad walo ko blac kar dega
ski.io.imshow(img_binary,)
```

```python
_, img_binary1 = cv2.threshold(img_gray,50,255,cv2.THRESH_BINARY_INV) #THRESH_BINARY_INV jo upar humne kiya hai uska inverse
#kardega
ski.io.imshow(img_binary1,)
```
